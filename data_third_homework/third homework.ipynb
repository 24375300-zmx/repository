```python
# 链式哈希表（使用 Python 内置真正的 hash 函数）
class HashTable:
    def __init__(self, size=10):
        self.size = size  # 哈希表容量
        self.table = [[] for _ in range(self.size)]  # 每个桶是一个链表

    # 真正的哈希函数：计算索引
    def hash(self, key):
        return hash(key) % self.size

    # 插入 / 更新
    def put(self, key, value):
        index = self.hash(key)
        bucket = self.table[index]

        # 如果键已存在，更新值
        for i in range(len(bucket)):
            k, v = bucket[i]
            if k == key:
                bucket[i] = (key, value)
                return

        # 键不存在，追加到桶里
        bucket.append((key, value))

    # 查询
    def get(self, key):
        index = self.hash(key)
        bucket = self.table[index]
        for k, v in bucket:
            if k == key:
                return v
        return None  # 没找到返回 None

    # 删除
    def delete(self, key):
        index = self.hash(key)
        bucket = self.table[index]
        for i in range(len(bucket)):
            k, v = bucket[i]
            if k == key:
                del bucket[i]
                return

    # 打印哈希表
    def show(self):
        for i in range(self.size):
            if self.table[i]:
                print(f"桶{i}: {self.table[i]}")
```

```python


# ==================== 测试（作业运行） ====================
if __name__ == "__main__":
    ht = HashTable(10)

    # 插入数据
    ht.put("Tom", 90)
    ht.put("Jerry", 85)
    ht.put("Alice", 95)
    ht.put("Bob", 88)

    print("=== 哈希表内容 ===")
    ht.show()

    # 查询
    print("\n查询 Tom 的成绩:", ht.get("Tom"))

    # 更新
    ht.put("Tom", 99)
    print("更新后 Tom 的成绩:", ht.get("Tom"))

    # 删除
    ht.delete("Bob")
    print("\n=== 删除 Bob 后的哈希表 ===")
    ht.show()
```
