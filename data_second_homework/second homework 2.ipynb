```python
from typing import Sequence, Any


class OrderedArray:
    def __init__(self, iterable: Sequence[Any]):
        """
        初始化有序数组（自动排序，确保初始数组有序）
        :param iterable: 可迭代对象（如列表、元组）
        """
        self.iterable = sorted(list(iterable))  # 确保初始数组有序
        self.len = len(self.iterable)

    def insert(self, element: Any) -> None:
        """
        线性查找插入元素，保持数组有序
        :param element: 待插入的元素
        """
        # 初始化插入位置为末尾（默认元素最大）
        insert_index = self.len

        # 线性遍历找第一个大于element的位置
        for index, value in enumerate(self.iterable):
            if element < value:
                insert_index = index
                break  # 找到位置后立即终止遍历，提升效率

        # 执行插入并更新数组长度
        self.iterable.insert(insert_index, element)
        self.len += 1

    def __str__(self) -> str:
        """格式化打印数组"""
        return str(self.iterable)
```

```python
# 测试示例
if __name__ == "__main__":
    # 初始化有序数组
    arr = OrderedArray([1, 4, 5, 19, 21])
    print("初始数组:", arr)  # 输出: [1, 4, 5, 19, 21]

    # 插入中间值
    arr.insert(7)
    print("插入7后:", arr)  # 输出: [1, 4, 5, 7, 19, 21]

    # 插入最小值
    arr.insert(0)
    print("插入0后:", arr)  # 输出: [0, 1, 4, 5, 7, 19, 21]

    # 插入最大值
    arr.insert(25)
    print("插入25后:", arr)  # 输出: [0, 1, 4, 5, 7, 19, 21, 25]
```
