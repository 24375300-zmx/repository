```python
class CollectionManager:
    """
    集合操作管理类，封装集合的查询、插入、删除功能
    """
    def __init__(self):
        # 初始化空集合，作为底层存储容器
        self.collection = set()

    def insert(self, element):
        """
        向集合中插入元素
        :param element: 要插入的元素（支持任意可哈希类型：数字、字符串、元组等）
        :return: 插入成功返回True，元素已存在返回False
        """
        if element in self.collection:
            print(f"提示：元素 {element} 已存在，无需重复插入")
            return False
        self.collection.add(element)
        print(f"元素 {element} 插入成功")
        return True

    def delete(self, element):
        """
        从集合中删除指定元素
        :param element: 要删除的元素
        :return: 删除成功返回True，元素不存在返回False
        """
        try:
            self.collection.remove(element)
            print(f"元素 {element} 删除成功")
            return True
        except KeyError:
            print(f"提示：元素 {element} 不存在，无法删除")
            return False

    def query(self, element):
        """
        查询元素是否存在于集合中
        :param element: 要查询的元素
        :return: 存在返回True，不存在返回False
        """
        exists = element in self.collection
        if exists:
            print(f"查询结果：元素 {element} 存在于集合中")
        else:
            print(f"查询结果：元素 {element} 不存在于集合中")
        return exists

    def get_all(self):
        """
        获取集合中所有元素（辅助方法，方便查看集合整体内容）
        :return: 集合的所有元素
        """
        return self.collection

    def clear(self):
        """清空集合（扩展方法，提升实用性）"""
        self.collection.clear()
        print("集合已清空")
```

```python
# ------------------- 测试示例 -------------------
if __name__ == "__main__":
    # 1. 创建集合管理实例
    cm = CollectionManager()

    # 2. 插入元素
    cm.insert(10)       # 插入成功
    cm.insert(20)       # 插入成功
    cm.insert(10)       # 元素已存在，插入失败

    # 3. 查询元素
    cm.query(10)        # 存在
    cm.query(30)        # 不存在

    # 4. 删除元素
    cm.delete(20)       # 删除成功
    cm.delete(30)       # 元素不存在，删除失败

    # 5. 查看当前集合所有元素
    print("当前集合元素：", cm.get_all())  # 输出 {10}

    # 6. 清空集合
    cm.clear()
    print("清空后集合元素：", cm.get_all())  # 输出 set()
```
