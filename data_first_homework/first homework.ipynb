```python
#作业代码
from typing import Iterable, Any, Optional


class ListOperation:
    def __init__(self, iterable: Iterable[Any]):
        """
        初始化列表操作类
        :param iterable: 可迭代对象，作为操作的基础列表
        """
        self.iterable = list(iterable)

    def insert(self, element: Any, index_to_insert: int) -> list[Any]:
        """
        向列表指定位置插入元素（复用之前的插入逻辑）
        :param element: 要插入的元素
        :param index_to_insert: 插入的索引位置
        :return: 插入后的列表
        """
        length = len(self.iterable)
        if index_to_insert < 0:
            index_to_insert = max(0, length + index_to_insert)
        elif index_to_insert > length:
            index_to_insert = length

        self.iterable.append("")  # 末尾预留空位
        # 与增加不同之处在于从后往前移动元素，避免覆盖
        for i in range(length, index_to_insert, -1):
            self.iterable[i] = self.iterable[i - 1]
        self.iterable[index_to_insert] = element
        return self.iterable

    def delete(self, index_to_delete: int) -> tuple[Optional[Any], list[Any]]:
        """
        删除列表指定位置的元素
        :param index_to_delete: 要删除的元素索引
        :return: (被删除的元素, 删除后的列表)
        """
        length = len(self.iterable)
        # 处理索引越界
        if length == 0:
            raise IndexError("无法从空列表中删除元素")
        if index_to_delete < 0 or index_to_delete >= length:
            raise IndexError("索引超出列表范围")

        # 保存要删除的元素
        deleted_element = self.iterable[index_to_delete]

        # 从删除位置的下一个元素开始，向前覆盖
        for i in range(index_to_delete, length - 1):
            self.iterable[i] = self.iterable[i + 1]

        # 移除最后一个冗余元素
        self.iterable.pop()

        return deleted_element, self.iterable
```

```python
#测试示例
if __name__ == "__main__":
    # 初始化列表
    fruits = ListOperation(["apple", "banana", "cherry", "orange", "mango"])
    print("初始列表:", fruits.iterable)  # 输出: ['apple', 'banana', 'cherry', 'orange', 'mango']

    # 测试删除操作：删除索引2的元素（cherry）
    deleted_elem, new_fruits = fruits.delete(2)
    print("被删除的元素:", deleted_elem)  # 输出: cherry
    print("删除后的列表:", new_fruits)  # 输出: ['apple', 'banana', 'orange', 'mango']

    # 测试删除第一个元素（apple）
    deleted_elem2, new_fruits2 = fruits.delete(0)
    print("被删除的元素:", deleted_elem2)  # 输出: apple
    print("删除后的列表:", new_fruits2)  # 输出: ['banana', 'orange', 'mango']
```
